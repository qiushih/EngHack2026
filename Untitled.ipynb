```python
import pandas as pd
```

```python
df =  pd.read_csv('Data sets - Spatial Data.csv')
```

```python
df = df.drop(0)
```

```python
df['X'] = pd.to_numeric(df['X'])
```

```python
df['Y'] = pd.to_numeric(df['Y'])
```

```python
df['Depth'] = pd.to_numeric(df['Depth'])
```

```python
df['Baseline CH4 Production Rate'] = pd.to_numeric(df['Baseline CH4 Production Rate'])
```

```python
df = df.drop(columns=['Solids'])
```

```python
df = df.drop(columns=['Clay'])
```

```python
df
```

```python
df['total'] = df['X'] * df['Y'] * df['Depth'] * df['Baseline CH4 Production Rate']
```

```python
df
```

```python
df = df.sort_values(by='total', ascending=False)
```

```python
df
```

```python
df = df.rename(columns= {'total': 'total mass /kg'})
```

```python
df
```

```python
df['Energy/MJ']= 55.5 * df['total mass /kg']
```

```python
df.head(100)
```

```python
df.to_csv('CH4_Emissions_Analysis.csv', index=False, encoding='utf_8_sig')
```

```python

df_style = df.sort_values(by='total mass /kg', ascending=False).head(100).style.format({
    'total mass /kg': '{:,.2f}',
    'Energy/MJ': '{:,.2f}',
    'Volume': '{:,.2e}'
}).background_gradient(subset=['Energy/MJ'], cmap='YlOrRd')

df_style
```

```python
def assign_zone(row):
    x, y = row['X'], row['Y']
    if 25 <= x <= 575 and 225 <= y <= 1425:
        return 'Zone 1: Western Sector'
    elif 1025 <= x <= 1575 and 1325 <= y <= 2075:
        return 'Zone 2: North-Central Hub'
    elif 625 <= x <= 1425 and 25 <= y <= 875:
        return 'Zone 3: South-Central Ridge'
    elif 2375 <= x <= 2475 and 1325 <= y <= 1475:
        return 'Zone 4: Eastern Outpost'
    else:
        return 'Other'

df['Zone'] = df.apply(assign_zone, axis=1)
zone_summary = df.groupby('Zone')['Energy/MJ'].sum().sort_values(ascending=False)
print(zone_summary)
```

```python
df.to_csv('CH4_Emissions_Analysis.csv', index=False, encoding='utf_8_sig')
```
