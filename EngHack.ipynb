```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['AirTemp'] = pd.to_numeric(df['AirTemp'])
df['SurfaceWaterTemp'] = pd.to_numeric(df['SurfaceWaterTemp'])
df['DeepWaterTemp'] = pd.to_numeric(df['DeepWaterTemp'])
df['pH'] = pd.to_numeric(df['pH'])
df['DOC'] = pd.to_numeric(df['DOC'])
df['Dissolved O2'] = pd.to_numeric(df['Dissolved O2'])
df['WindSpeed'] = pd.to_numeric(df['WindSpeed'])
df['Ice Cover Fraction'] = pd.to_numeric(df['Ice Cover Fraction'])
df['Measured CH4 Flux'] = pd.to_numeric(df['Measured CH4 Flux'])
```

```python
df.info()
```

```python
df
```

```python
correlation_AirTemp = df['AirTemp'].corr(df['Measured CH4 Flux'])
correlation_pH = df['pH'].corr(df['Measured CH4 Flux'])
correlation_DOC = df['DOC'].corr(df['Measured CH4 Flux'])
correlation_Dissolved_O2 = df['Dissolved O2'].corr(df['Measured CH4 Flux'])
correlation_WindSpeed = df['WindSpeed'].corr(df['Measured CH4 Flux'])
correlation_DeepWaterTemp = df['DeepWaterTemp'].corr(df['Measured CH4 Flux'])
```
